```python
#Helper Function

import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_modified = movies.copy()

movies_modified['keywords'] = movies_modified['keywords'].apply(convert)


# movies['keywords'] = movies['keywords'].apply(convert)
```

```python

def convert3(obj):
    L=[]
    counter = 0;
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    

movies_modified['cast'] = movies_modified['cast'].apply(convert3)
```

```python
movies_modified['genres'] = movies_modified['genres'].apply(convert)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L      


movies_modified['crew'] = movies_modified['crew'].apply(fetch_director)
```

```python
#convert the string into list using lambda function


movies_modified['overview'] = movies_modified['overview'].apply(lambda x:x.split() if isinstance(x,str) else [])


# todo -> fill NA values.
```

```python
movies_modified['genres'] = movies_modified['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_modified['keywords'] = movies_modified['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_modified['cast'] = movies_modified['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_modified['crew'] = movies_modified['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_modified['tags'] = movies_modified['overview'] + movies_modified['genres'] + movies_modified['keywords'] + movies_modified['cast'] + movies_modified['crew']
```

```python
# because of same word appear multiple time in our dataset we can use "Stemming" ,ant for use stemming we import the library call "NLTK"
#(Natural Language Toolkit)

!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem) 
```

```python
# ------------------- vectorization process starts --------------------

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```
